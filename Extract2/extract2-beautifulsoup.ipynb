```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('http://quotes.toscrape.com/')
```

```python
soup = BeautifulSoup(response.text, 'lxml')
```

```python
print(soup.prettify())
```

```python
quotes = soup.find_all('span', class_='text')
authors = soup.find_all('small', class_='author')
tags = soup.find_all('div', class_ ='tags')
for i in range(len(quotes)):
    print(quotes[i].text)
    print(authors[i].text, '\n')
    quote_tags = tags[i].find_all('a', class_='tag')
    
    for j in quote_tags:
        print(j.text)
    print()

    
    
    
    
    
    
    
```

```python
file = open('quotes.txt', 'w')
quotes = soup.find_all('span', class_='text')
authors = soup.find_all('small', class_='author')
tags = soup.find_all('div', class_ ='tags')
for i in range(len(quotes)):
    file.write(quotes[i].text)
    file.write('\n')
    file.write(authors[i].text)
    file.write('\n')
    quote_tags = tags[i].find_all('a', class_='tag')
    
    for j in quote_tags:
        file.write(j.text)
        file.write('\n')
    file.write('\n')
file.flush()
file.close()
```
