```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = requests.get('http://en.wikipedia.org/wiki/List_of_A_Song_of_Ice_and_Fire_characters')
soup = BeautifulSoup(url.content, 'lxml')
```

```python
soup.find_all('h3')[-4].span.get_text()
```

```python
soup.find_all('span', class_="mw-headline")[-10]
```

```python
print(soup.prettify()[:100000])
```

```python
soup.find_all('li', class_= "toclevel-2")[-4]
```

```python

lists = []
items = soup.find_all('li', class_= "toclevel-2")[:-3]
for i in items:
    lists.append(i.find('span', class_="toctext").text)
```

```python
lists
```

```python
houses = []
for name in lists:
    houses.append(name.split()[-1])
```

```python
houses
```

```python
houses = pd.Series(houses)
```

```python
houses.value_counts()
```

```python
houses.value_counts().to_frame()
```
