```python
import requests
from bs4 import BeautifulSoup
from IPython.display import Image
import matplotlib.pyplot as plt
```

```python
url = requests.get('https://scrapingclub.com/exercise/list_basic/')
soup = BeautifulSoup(url.content, 'lxml')
```

```python
print(soup.prettify())
```

```python
products = soup.find_all('div', 'col-lg-4 col-md-6 mb-4')
```

```python
images = []
count = 1
for i in range(len(products)):
    item_name = products[i].h4.text.strip()
    item_price = products[i].h5.text
    images.append('https://scrapingclub.com'+products[i].img['src'])
    print('{}) {} : {}'.format(count, item_name, item_price))
    count +=1
    
```
